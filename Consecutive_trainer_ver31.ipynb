```python
def MaxMinStandardize(data):
  
  
  #print(data)
  
  data_std = (data - data.min()) / (data.max() - data.min())     
        
    
  return data_std
```

```python
def TSprocessDiff(tsdata_set, n_in = 10, n_out=5, step=1):

  datasets       = []  

  for tsdata in tsdata_set:   # for each dataset 
        
        dataset_ = []
        
        # max min standardize
        
        tsdata_std = MaxMinStandardize(tsdata)
        
        # take difference from last period
        tsdata_diff = tsdata_std.diff()
        
        tsdata_diff = tsdata_diff.dropna().reset_index()
        #tsdata_diff = tsdata_diff.reset_index()
  
        
        # unfold 
        unfolded_tsdata  =  torch.FloatTensor(tsdata_diff['x']).unfold(0, n_in + n_out, step)
       
       
        for k in range(len(unfolded_tsdata)):        
          
          dataset_.append([unfolded_tsdata[k, :n_in],  unfolded_tsdata[k, n_in:]])    # [X (data),y (target)] 
            
        datasets.append(dataset_)

  return datasets
```

```python

# Preparing dataset 

from torch.utils.data.dataset import Dataset

class LSTM_dataset(Dataset):
  
  def __init__(self, tsdata_set):
    
    self.output_dataset = tsdata_set
    
  def __getitem__(self,index):
    
    X = torch.FloatTensor(self.output_dataset[index][0])
    y = torch.FloatTensor(self.output_dataset[index][1])
    
    return X, y
  
  def __len__(self):
    
    length = len(self.output_dataset)
    
    return length
  
```

```python
# LSTM model 

import torch
from torch import nn
from torch.autograd import Variable


class LSTM(nn.Module):
    def __init__(self, hidden_size=64, num_timeseries=1, dropout=0.1, n_layers=1, time=10):
          
        super(LSTM, self).__init__()
        self.n_layers = n_layers
        self.hidden_size = hidden_size
        self.LSTM = nn.LSTM(
            input_size=num_timeseries,
            hidden_size=self.hidden_size,
            num_layers=n_layers,
            dropout=dropout,
            bidirectional=False
        )

        self.fc1 = nn.Linear(self.hidden_size, self.hidden_size)
        self.fc2 = nn.Linear(self.hidden_size, time)
        self.relu = nn.ReLU()
        
        self.h = None
        self.c = None 
         
        self.reset = True
        

    def forward(self, x, reset=False):

        batch_size = x.size()[1]
        
        if self.reset == True: 
          reset = True 
          self.rest = False 
        
        # LSTM
        if reset == True: 
         
          h0 = Variable(torch.zeros(self.n_layers, batch_size, self.hidden_size)).cuda()
          c0 = Variable(torch.zeros(self.n_layers, batch_size, self.hidden_size)).cuda()
        else:
          h0 = self.h
          c0 = self.c
        #c = c.cuda()
        #h = h.cuda()
        
        outputs, (h, c) = self.LSTM(x, (h0, c0))       
        out = self.relu(outputs)
        
        out = outputs[-1]  # last prediction

        # Fully connected 
        out = self.fc1(out)
             
        out = self.relu(out)
        out = self.fc2(out)
        
        self.h = h
        self.c = c
                    
        return out
      
    def reset_hidden(self):
        self.reset = True 
```
